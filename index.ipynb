```python
from main import *
import pandas
import numpy
import ast

showMenu()
menuInput = input('Option : ')

while True:
    print()
    
    if menuInput == '0':
        break
        
    elif menuInput == '1':
        menuPriceDataCustomer = []
        menuDataCustomer = []
        menuPriceDataArray = [0]
        menuDataArray = ['0']
        
        customerName = input('Costumer\'s Name : ')
        menuDatabase = open('menu.json', 'r')
        menuData = ast.literal_eval(menuDatabase.read())
        for menu in menuData['Price']:
            menuPriceDataArray.append(menu)
        for menu in menuData['Menu']:
            menuDataArray.append(menu)
        menuDatabase.close()
        
        while True:
            print()
            print(pandas.DataFrame(data = menuData, index = [*range(1, len(menuPriceDataArray))]))
            customerOrder = int(input('Costumer\'s Order : '))
            
            if customerOrder == 0:
                break
            
            else:
                try:
                    menuDataCustomer.append(menuDataArray[customerOrder])
                    menuPriceDataCustomer.append(menuPriceDataArray[customerOrder])
                    
                except:
                    print('Data Not Found With That ID')
            
        try:
            customerDatabase = open('database.json', 'r')
            customerData = ast.literal_eval(customerDatabase.read())
            customerDatabase.close()
            
            customerDatabase = open('database.json', 'w')
            if getTodayDate() in customerData:
                # Check If Costumer Already Exist
                customerData[getTodayDate()].append({
                    "name" : customerName,
                    "order" : menuDataCustomer,
                    "total" : sum(menuPriceDataCustomer)
                })
            else:
                customerData[getTodayDate()] = []
                customerData[getTodayDate()].append({
                    "name" : customerName,
                    "order" : menuDataCustomer,
                    "total" : sum(menuPriceDataCustomer)
                })
            customerDatabase.write(str(customerData))
            customerDatabase.close()
            
        except:
            print('Data Not Found')
    
    elif menuInput == '2':
        
        while True:
            showReportMenu()
            menuInputReport = input('Option : ')
            
            if menuInputReport == '0':
                break
            
            elif menuInputReport == '1':
                database = open('database.json',  'r')
                
                database.close()
    
    print()
    showMenu()
    menuInput = input('Option : ')
```
