```python
from main import *
import pandas, numpy, ast, calendar, numpy
from datetime import datetime, timedelta

showMenu()
menuInput = input('Option : ')

while True:
    print()
    
    if menuInput == '0':
        break
        
    elif menuInput == '1':
        menuPriceDataCustomer = []
        menuDataCustomer = []
        menuPriceDataArray = [0]
        menuDataArray = ['0']
        
        customerName = input('Costumer\'s Name : ')
        menuDatabase = open('menu.json', 'r')
        menuData = ast.literal_eval(menuDatabase.read())
        for menu in menuData['Price']:
            menuPriceDataArray.append(menu)
        for menu in menuData['Menu']:
            menuDataArray.append(menu)
        menuDatabase.close()
        
        while True:
            print()
            print(pandas.DataFrame(data = menuData, index = numpy.arange(1, len(menuPriceDataArray))))
            customerOrder = int(input('Costumer\'s Order (Enter 0 If Done) : '))
            
            if customerOrder == 0:
                break
            
            else:
                try:
                    menuDataCustomer.append(menuDataArray[customerOrder])
                    menuPriceDataCustomer.append(menuPriceDataArray[customerOrder])
                except:
                    print('Data Not Found With That ID')
        try:
            customerDatabase = open('database.json', 'r')
            customerData = ast.literal_eval(customerDatabase.read())
            customerDatabase.close()

            customerDatabase = open('database.json', 'w')
            if getTodayDate() in customerData:

                try:
                    getIndexCustomerName = [x['name'] for x in customerData[getTodayDate()]].index(customerName)
                except:
                    getIndexCustomerName = -1

                if getIndexCustomerName != -1:

                    customerData[getTodayDate()][getIndexCustomerName]['order'] += menuDataCustomer
                    customerData[getTodayDate()][getIndexCustomerName]['total'] += sum(menuPriceDataCustomer)

                else:
                    customerData[getTodayDate()].append({
                        "name" : customerName,
                        "order" : menuDataCustomer,
                        "total" : sum(menuPriceDataCustomer)
                    })
            else:
                customerData[getTodayDate()] = []
                customerData[getTodayDate()].append({
                    "name" : customerName,
                    "order" : menuDataCustomer,
                    "total" : sum(menuPriceDataCustomer)
                })
            customerDatabase.write(str(customerData))
            customerDatabase.close()
            print('Data Order Added')
        except:
            print('Data Not Found')
    
    elif menuInput == '2':
        
        while True:
            print()
            showReportMenu()
            menuInputReport = input('Option : ')
            
            if menuInputReport == '0':
                break
            
            elif menuInputReport == '1':
                print()
                try:
                    database = open('database.json',  'r')
                    databaseData = ast.literal_eval(database.read())
                    total = 0
                    for i in databaseData[getTodayDate()]:
                        total += i['total']
                    print(pandas.DataFrame(data = databaseData[getTodayDate()], index = numpy.arange(1, len(databaseData[getTodayDate()]) + 1)))
                    showFooter('=', 80)
                except:
                    print('Data Not Found')
                database.close()
            
            elif menuInputReport == '2':
                print()
                try:
                    database = open('database.json',  'r')
                    databaseData = ast.literal_eval(database.read())
                    now = datetime.now()
                    monday = now - timedelta(days = now.weekday())
                    dataReportPerWeek = []
                    dataTotalPerDay = []
                    totalPerWeek = 0
                    
                    for i in range(0, 7):
                        totalPerDay = 0
                        date = (monday + timedelta(days = i)).strftime('%d/%m/%Y')
                        try:
                            if date in databaseData:
                                for a in databaseData[date]:
                                    totalPerDay += a['total']
                                    a['date'] = date
                                    dataReportPerWeek.append(a)
                            dataTotalPerDay.append(totalPerDay)
                            totalPerWeek += totalPerDay
                        except:
                            print('Data Not Found')
                    print()
                    print(pandas.DataFrame(data = dataReportPerWeek, index = numpy.arange(1, len(dataReportPerWeek) + 1), columns = ['date', 'name', 'order', 'total']))
                    print()
                    print(pandas.DataFrame(data = dataTotalPerDay, columns = ['Total'], index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']))
                    print()
                    print('Total This Week : ' + str(totalPerWeek))
                    showFooter('=', 80)
                except:
                    print('Something\'s Went Wrong!')
                database.close()
                
            elif menuInputReport == '3':
                print()
                try:
                    database = open('database.json',  'r')
                    databaseData = ast.literal_eval(database.read())
                    now = datetime.now()
                    dataReportPerMonth = []
                    totalPerMonth = 0

                    for i in range(1, int((now.replace(day = calendar.monthrange(now.year, now.month)[1])).strftime('%d')) + 1):
                        date = (datetime.today().replace(day = i)).strftime('%d/%m/%Y')
                        try:
                            if date in databaseData:
                                totalPerDay = 0
                                for a in databaseData[date]:
                                    totalPerMonth += a['total']
                                    a['date'] = date
                                    dataReportPerMonth.append(a)
                        except:
                            print('Data Not Found')
                    print(pandas.DataFrame(data = dataReportPerMonth, index = numpy.arange(1, len(dataReportPerMonth) + 1), columns = ['date', 'name', 'order', 'total']))
                    print()
                    print('Total This Month : ' + str(totalPerMonth))
                    showFooter('=', 80)
                except:
                    print('Data Not Found')
                database.close()
            
            elif menuInputReport == '4':
                print()
                try:
                    database = open('database.json',  'r')
                    databaseData = ast.literal_eval(database.read())
                    now = datetime.now()
                    thisYearData = []
                    total = 0

                    for i in databaseData:
                        try:
                            getYear = i.split('/')[2]
                            if getYear == str(now.year):
                                for a in databaseData[i]:
                                    a['date'] = i
                                    total += a['total']
                                thisYearData.extend(databaseData[i])
                        except:
                            False
                    print(pandas.DataFrame(data = thisYearData, index = numpy.arange(1, len(thisYearData) + 1), columns = ['date', 'name', 'order', 'total']))
                    print()
                    print('Total This Year = ' + str(total))
                    showFooter('=', 80)
                except:
                    print('Data Not Found')
                database.close()
            
            elif menuInputReport == '5':
                print()
                try:
                    database = open('database.json',  'r')
                    databaseData = ast.literal_eval(database.read())
                    customerName = input('Customer\'s Name : ')
                    customerReportData = []
                    total = 0

                    for i in databaseData:
                        try:
                            for a in databaseData[i]:
                                try:
                                    if a['name'] == customerName:
                                        a['date'] = i
                                        total += a['total']
                                        customerReportData.append(a)
                                except:
                                    False
                        except:
                            False
                    if len(customerReportData) == 0:
                        print('Data Not Found')
                    else:
                        print(pandas.DataFrame(data = customerReportData, columns = ['date', 'order', 'total'], index = numpy.arange(1, len(customerReportData) + 1)))
                        print()
                        print('Total Spent : ' + str(total))
                        showFooter('=', 80)
                except:
                    print('Data Not Found')
                database.close()
            
            elif menuInputReport == '6':
                print()
                try:
                    database = open('database.json',  'r')
                    databaseData = ast.literal_eval(database.read())
                    inputMenu = input('Menu\'s Name : ')
                    dataReportMenu = []
                    total = 0

                    for i in databaseData:
                        try:
                            for a in databaseData[i]:
                                if inputMenu in a['order']:
                                    a['date'] = i
                                    total += a['total']
                                    dataReportMenu.append(a)
                        except:
                            False
                    print(pandas.DataFrame(data = dataReportMenu, index = numpy.arange(1, len(dataReportMenu) + 1), columns = ['date', 'name', 'order', 'total']))
                    print()
                    print('Total : ' + str(total))
                    showFooter('=', 80)
                except:
                    print('Data Not Found')
                database.close()
            
            elif menuInputReport == '7':
                print()
                try:
                    database = open('database.json',  'r')
                    databaseData = ast.literal_eval(database.read())
                    dataReport = []
                    total = 0

                    for i in databaseData:
                        try:
                            for a in databaseData[i]:
                                a['date'] = i
                                total += a['total']
                                dataReport.append(a)
                        except:
                            False
                    print(pandas.DataFrame(data = dataReport, index = numpy.arange(1, len(dataReport) + 1), columns = ['date', 'name', 'order', 'total']))
                    print()
                    print('Total : ' + str(total))
                    showFooter('=', 80)
                except:
                    print('Data Not Found')
                database.close()
                
    print()
    showMenu()
    menuInput = input('Option : ')
```
